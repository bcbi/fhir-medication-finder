```python
import requests
import os
from flask import Flask, request, render_template
from dotenv import load_dotenv
```

```python
s = requests.Session()
```

```python
patient_id = 1

load_dotenv()

username = os.getenv("FHIR_USERNAME")
password = os.getenv("FHIR_PASSWORD")


credentials = (username, password)

r = s.get("http://pwebmedcit.services.brown.edu:8081/fhir" + f"/MedicationRequest?patient={patient_id}", auth=credentials)
r
```

```python
patient = s.get(f"http://pwebmedcit.services.brown.edu:8081/fhir/MedicationRequest?patient={patient_id}", headers = {'Accept': 'application/fhir/json'}, auth=credentials) 
patient
```
